```python
import pandas as pd
import csv
import glob
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# load csvs
csv_files_path = "./data/*.csv"
csv_files = glob.glob(csv_files_path)
dfs = []
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df["date"] = "01-17-2024"
    df["cycle"] = csv_file.split("_")[-1].replace(".csv", "")
    dfs.append(df)


combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df)
df=combined_df
```

```python
combined_df.to_csv("combined.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["Loc Name"], df["DC"], color="skyblue")
plt.title("DC Values for Locations")
plt.xlabel("Location Name")
plt.ylabel("DC Values")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for 'OPC' and 'TSQ'
plt.figure(figsize=(10, 6))
plt.scatter(df["OPC"], df["TSQ"], color="orange")
plt.title("Scatter Plot of OPC vs. TSQ")
plt.xlabel("OPC Values")
plt.ylabel("TSQ Values")
plt.tight_layout()
plt.show()
```

```python
df.groupby("cycle")["TSQ"].mean().reset_index()
```

```python
df["cycle"] = df["cycle"].astype("category").cat.reorder_categories(["Timely", "Evening", "Intraday 1", "Intraday 2", "Intraday 3", "Final"])
grouped_df = df.groupby("cycle")["TSQ"].mean().reset_index()

# Line plot for 'OPC' grouped by 'cycle'
plt.figure(figsize=(10, 6))
plt.plot(
    grouped_df["cycle"],
    grouped_df["TSQ"],
    marker="o",
    color="green",
    linestyle="-",
    linewidth=2,
    markersize=8,
)
plt.title("Line Plot of TSQ Grouped by Cycle")
plt.xlabel("Cycle")
plt.ylabel("Mean TSQ Values")
plt.grid(True)
plt.show()
```
